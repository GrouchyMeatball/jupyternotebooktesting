```python
import csv, sqlite3
import pandas
#con = sqlite3.connect("SQLiteMagic.db")
#cur = con.cursor()
chicago_socioeconomic_data = pandas.read_csv('https://data.cityofchicago.org/resource/jcxq-k9xf.csv')
chicago_public_schools = pandas.read_csv('https://data.cityofchicago.org/resource/9xs2-f89t.csv')
chicago_crime_data = pandas.read_csv('https://data.cityofchicago.org/resource/ijzp-q8t2.csv')
%sql sqlite:///SQLiteMagic.db
%sql PERSIST chicago_socioeconomic_data
%sql PERSIST chicago_public_schools
%sql PERSIST chicago_crime_data
```

## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
%sql select count(*) from chicago_crime_data
#The total number of crimes recorded in the Crime table is 1000
```

### Problem 2

##### List community areas with per capita income less than 11000.


```python
%sql select community_area_name from chicago_socioeconomic_data where per_capita_income_ < 11000;
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)


```python
%%sql 

select * from chicago_crime_data where (description like '%minor%' and description not like '%child%' and primary_type not like '%child%');
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
%sql select * from chicago_crime_data where primary_type like '%KIDNAPPING%'
```

### Problem 5

##### What kinds of crimes were recorded at schools?


```python
%sql select * from chicago_crime_data where location_description like '%SCHOOL%'
# The kinds of crimes were recorded at schools are criminal damage,motor vehicle theft, battery, assault, other offense and sexual offense.
```

### Problem 6

##### List the average safety score for each type of school.


```python
%sql select elementary_or_high_school, avg(safety_score) from CHICAGO_PUBLIC_SCHOOLS group by elementary_or_high_school
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
%sql select * from chicago_socioeconomic_data order by percent_households_below_poverty desc limit 5;
```

### Problem 8

##### Which community area is most crime prone?


```python
%%sql 
select community_area, count(*) as crime_prone_count
from chicago_crime_data 
group by community_area
order by crime_prone_count desc limit 1

#Community area 25 is most crime prone
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
%sql select * from chicago_socioeconomic_data order by hardship_index desc limit 1

# The name of the community area with highest hardship index is Riverdale
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
%%sql 
select community_area, count(*) as crime_prone_count
from chicago_crime_data 
group by community_area
order by crime_prone_count desc limit 1
```
